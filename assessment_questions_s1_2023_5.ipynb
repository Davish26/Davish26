````python
#The os module allows us to gain access to our paths for the file
import os
#This will help the user to get to know which option she can choose.
print("Option 1 : Counting the number of lines")
print("Option 2 : Counting total number of words")
print("Option 3 : Counting the total number of characters, with or without whitespace")
print("Option 4 : Identifying the frequency of each word")
print("Option 5 : Identifying the top 5 frequent words")

#The function count_lines allows us to open and read a file, then outputs the number of lines
def count_lines(file_path):
    with open(file_path, 'r') as file:
#For each line, it increaments by 1
        count_line = sum(1 for line in file)
    return count_line

#The function count_words open and read the designated file
def count_words(file_path):
    with open(file_path, 'r') as file:
#Splits each lines into words and count the number of words using len
        count_word = sum(len(line.split()) for line in file)
    return count_word

#The function count_characters seperates each words and counts the number of characters with or without space depending the user
def count_characters(file_path, include_whitespace=True):
    with open(file_path, 'r') as file:
        content = file.read()
#Asks the user if he wants to count space or not
        whitespace = (input("Do you want to include white space? "))
        whitespace = whitespace.lower()
        if whitespace == "yes":
            count_character = len(content)
        elif whitespace == "no":
            content = content.translate(str.maketrans('', '', whitespace))
            count_character = len(content)
    return count_character

#The function word_frequency counts the number of times each word have appeared
def word_frequency(file_path):
    word_count = {}
    with open(file_path, 'r') as file:
        for line in file:
            words = line.lower().split()
            for word in words:
#For every similar word, it adds one to the count
                word_count[word] = word_count.get(word, 0) + 1
    return word_count

#The function top_5 outputs us with the 5 most frequent word in the file
def top_5(file_path):
    word_count = word_frequency(file_path)
    top_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)[:5]
    return top_words

#The function analysis allows us to make a choice between the 5 different options.
def analysis(file_path, option):
    if option == '1':
        line_count = count_lines(file_path)
        print("Number of lines: ", line_count)
    elif option == '2':
        word_count = count_words(file_path)
        print("Number of words: ", word_count)
    elif option == '3':
          count_character = count_characters(file_path)
          print("Number of characters: ", count_character)
    elif option == '4':
        word_count = word_frequency(file_path)
        for word, count in word_count.items():
            print("Word frequency: ", word, count)
    elif option == '5':
        top_words = top_5(file_path)
        print("Top 5 words: ")
        for word, count in top_words:
          print(word, count)
#Incase the user inputs an invalid option number. He will get an output message, asking him to try again.
    else:
        print("You entered an invalid option. Please try again.")
#Allows the user to input path to file
file_path = input("Enter the path to the text file: ")
if not os.path.isfile(file_path):
    print("File not found. Please enter a valid file path.")
#Incase the user inputs an invalid file path, he will be asked to input an other valid file path

#Allows the user to input his option number
option = input("Enter your option: ")
analysis(file_path, option)
````

````python
import newspaper
from newspaper import Article

def article_details(url):
    article = Article(url)
#Download the url content
    article.download()
#Extract content of the url once downloaded
    article.parse()
    return {"url": url, "title": article.title, "text": article.text}

def articles_summary(urls):
    articles = []
    for url in urls:
        try:
            article_info = article_details(url)
            articles.append(article_info)
        except Exception as e:
            print("Error processing URL:", url)
    return articles

url_list = [
    "https://www.theguardian.com/world/2023/jun/12/ukraine-mayor-sees-opportunity-chaos-nova-kakhovka-collapse-local-history-dnipro-river",
    "https://www.theguardian.com/football/2023/jun/12/sweden-players-had-to-show-their-genitalia-at-2011-womens-world-cup-says-nilla-fischer"
]

list = articles_summary(url_list)
for article in list:
    print("URL:", article["url"])
    print("Title:", article["title"])
    print("Text:", article["text"])
````
